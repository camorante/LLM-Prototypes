<h2 style="color:Blue">General Functions</h2>

```python
def load_image(path, scale=1.2):
    imagen = plt.imread(path)
    plt.figure(figsize=(imagen.shape[1] * scale / 100, imagen.shape[0] * scale / 100))
    plt.imshow(imagen)
    plt.show()
```
